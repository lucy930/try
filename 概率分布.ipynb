# python 实现伯努利分布

```python
import scipy
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''定义随机变量：1次抛硬币成功指正面朝上记为1，失败反面朝上记为0'''
x=np.arange(0,2,1)
#伯努利分布：Bernoulli Distribution
#概率质量函数PMF
#概率密度函数PDF
p=0.5
plist=stats.bernoulli.pmf(x,p)
plt.plot(x,plist,marker='o',linestyle='None')
plt.vlines(x,0,plist)
#vline(x坐标值，y坐标最小值，y坐标值最大值，在这里传入的x是一个数组，竖直线y坐标最小值是0，y坐标值最大值是对应plist中的值)
plt.xlabel('随机变量：抛硬币1次')
plt.ylabel('概率')
plt.title('伯努利分布：p=%.2f'%p)
plt.show()
```

# Python实现二项分布

```python
#定义随机变量，抛5次硬币，正面朝上的次数
n=5
p=0.5
x=np.arange(0,n+1,1)
plist=stats.binom.pmf(x,n,p)
#Binomial Distribution二项分布
plt.plot(x,plist,marker='o',linestyle='None')
plt.vlines(x,0,plist)
plt.xlabel('随机变量：抛硬币正面朝上次数')
plt.ylabel('概率')
plt.title('二项分布：n=%i,p=%.2f'%(n,p))
plt.show()
```

# python 实现几何分布

```python
#从产品中随机选取样品进行检验，次品率为0.6，共抽5次求第i次抽到次品的概率。
k=6
p=0.5
x=np.arange(1,k+1,1)
#几何分布 Geometric Distribution
plist=stats.geom.pmf(x,p)
plt.plot(x,plist,marker='o',linestyle='None')
plt.vlines(x,0,plist)
plt.xlabel('随机变量：第i次抽到次品的概率')
plt.ylabel('概率')
plt.title('几何分布:p=%.2f'%p)
plt.show()
```

# python实现泊松分布

```python
#在一公里内公路损坏n段的概率,假设每公里平均有2次损坏，求某一公里内有4次损坏的概率
u=2
k=4
x=np.arange(0,k+1,1)
plist=stats.poisson.pmf(x,u)
plt.plot(x,plist,marker='o',linestyle='None')
plt.vlines(x,0,plist)
plt.xlabel('随机变量：某公里内有k次损坏的概率')
plt.ylabel('概率')
plt.title('泊松分布：平均值u=%i'%u)
plt.show()
```
